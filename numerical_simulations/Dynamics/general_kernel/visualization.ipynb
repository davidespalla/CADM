# General Kernel

## GAUSSIAN DERIVATIVE

```python
pfc=load("1D_GAUSSIAN//pfc.npy")
Vs=load("1D_GAUSSIAN//Vdynamics.npy")
```

```python
t=40
Dt=5

color=cm.Blues(np.linspace(0.5,1,3))

fig, ax = plt.subplots(figsize=(6,6))
plot(pfc,Vs[t-Dt],label=r"$t_{1}$",c=color[0], linestyle="-",linewidth=3)
plot(pfc,Vs[t],label=r"$t_{2}$",c=color[1], linestyle="-",linewidth=3)
plot(pfc,Vs[t+Dt],label=r"$t_{3}$",c=color[2], linestyle="-",linewidth=3)
xlabel("encoded feature $x$", fontsize=20)
ylabel("activity level $V(x)$", fontsize=20)
ax.tick_params(axis='both', which='major', labelsize=15)
#title(r"$K(d)=e^{-|d|}+\gamma d \mathcal{S}(d) $",fontsize=25)
savefig("Fig4_gauss.svg")
#legend()
```
