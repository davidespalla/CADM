```python
%pylab inline
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.markers import MarkerStyle
```

```python
def com(V,pfc):
    x=sum(V*pfc[0].T[0])/len(V)
    y=sum(V*pfc[0].T[1])/len(V)
    return [x,y]
```

# Crossing point analysis

```python
figure(figsize=(20,15))
gammas=load("crossing_study/gammas.npy")
xlim([0,10])
ylim([0,10])
for i in range(20):
    pfc=load("crossing_study/gamma_"+str(i+1)+"/pfc.npy")
    Vs=load("crossing_study/gamma_"+str(i+1)+"/Vdynamics.npy")
    xc=[]
    yc=[]
    for j in range(len(Vs)):
        xc.append(com(Vs[j],pfc)[0])
        yc.append(com(Vs[j],pfc)[1])
    ax=subplot(4,5,i+1)
    ax.scatter(xc,yc)
    ax.set_title("gamma="+str(gammas[i]))
    ax.set_xlim([0,10])
    ax.set_ylim([0,10])
    
tight_layout()
    
```

```python
figure(figsize=(4,4))
gammas=load("crossing_study/gammas.npy")
xlim([0,10])
ylim([0,10])
indexes=[0,3,6]
pfc=load("crossing_study/gamma_"+str(6)+"/pfc.npy")
Vs=load("crossing_study/gamma_"+str(6)+"/Vdynamics.npy")
endtime=12
xc=[]
yc=[]
for j in range(endtime):
    xc.append(com(Vs[j],pfc)[0])
    yc.append(com(Vs[j],pfc)[1])
    
plot(xc,yc,linewidth=5,label=r"$\gamma={:10.2f}$".format(gammas[1]))
arrow(5,0.5,0,8.5,head_width=0.3,color="k",linestyle="--")

pfc=load("crossing_study/gamma_"+str(15)+"/pfc.npy")
Vs=load("crossing_study/gamma_"+str(15)+"/Vdynamics.npy")
endtime=7
xc=[]
yc=[]
for j in range(endtime):
    xc.append(com(Vs[j],pfc)[0])
    yc.append(com(Vs[j],pfc)[1])
    
plot(xc,yc,linewidth=5,label=r"$\gamma={:10.2f}$".format(gammas[15]))

arrow(0.5,5,8.5,0,head_width=0.3,color="k",linestyle="--")
xlabel(r"feature 1",fontsize=15)
ylabel(r"feature 2",fontsize=15)

xticks(fontsize=10)
yticks(fontsize=10)

legend(fontsize=12,loc='lower left')    
tight_layout()
savefig("corssing_examples.svg")
```
